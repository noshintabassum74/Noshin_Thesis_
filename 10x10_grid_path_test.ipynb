# Signal Strength Functions

```python
def Prof_calculate_signal_strength(map_x_coordinate, map_y_coordinate, point_x_coordinate, point_y_coordinate):
  point_one = np.array((map_x_coordinate, map_y_coordinate))
  point_two = np.array((point_x_coordinate, point_y_coordinate))
  euclidean_distance = np.linalg.norm(point_one - point_two)
  # n = 3
  # return 1/(euclidean_distance**n+0.0001)
  frequency = 2412 

  signal_strength = ((20 * np.log(euclidean_distance)) / (np.log(10))) + 20 * np.log10(frequency) - 27.55
  return signal_strength
```

```python
def sig_strength_access_points(coords):
  x,y = coords
  sig_strength = []
  for row in grid_map:
    for accessPoint in row:
      # print(accessPoint)
      strength = Prof_calculate_signal_strength(accessPoint[0], accessPoint[1], x, y)
      # print(accessPoint, x, y, strength)
      sig_strength.append(strength)
  return sig_strength
```

# Compute Distance Matrix

```python
def compute_distance_matrix(trajectories, method="dtw"):
        """
        :param method: "Frechet" or "Area"
        """
        n = len(trajectories)
        dist_m = np.zeros((n, n))
        for i in range(n - 1):
            p = trajectories[i]
            for j in range(i + 1, n):
                q = trajectories[j]
                if method == "Frechet":
                    dist_m[i, j] = similaritymeasures.frechet_dist(p, q)
                else:
                    dist_m[i, j], d = similaritymeasures.dtw(p, q)
                dist_m[j, i] = dist_m[i, j]
        return dist_m
# print(Data_Scaled)
Sig_strength_score_mse =  compute_distance_matrix(np.array(Sig_strength_score))
# Sig_strength_score_mse =  compute_distance_matrix(Sig_strength_score)
print(np.amax(Sig_strength_score_mse))
```
